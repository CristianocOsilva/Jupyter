```python
import pandas as pd
from pandas_datareader import data as web
import matplotlib.pyplot as plt

# vamos pegar cotaÃ§Ã£o do Indice e de Petrobras

#indice
df = web.DataReader(f'^BVSP', data_source='yahoo', start=f'03-01-2020', end='05-01-2021')
display(df)
df["Adj Close"].plot(figsize=(10, 5))
plt.show()

#BBAS3
df = web.DataReader(f'BBAS3.SA', data_source='yahoo', start=f'02-20-2020', end='02-20-2021')
display(df)
df["Adj Close"].plot(figsize=(10, 5))
plt.show()
```

```python
data_inicial = "01/01/2020"
data_final = "01/01/2021"

empresas_df = pd.read_excel("Empresas.xlsx")
display(empresas_df)

for empresa in empresas_df['Empresas']: 
    print(f"{empresa}:")
    df = web.DataReader(f'{empresa}.SA', data_source='yahoo', start=data_inicial, end=data_final)
    display(df)
    df["Adj Close"].plot(figsize=(15, 10))
    plt.show()
```
